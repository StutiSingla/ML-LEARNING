```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv ("titanic.csv")
```

```python
df.head()
```

**CATEGORICAL DATA**

On survived column

**COUNTPLOT** - It counts the frequency of each unique value.

```python
sns.countplot(x='Survived', data=df)
#x="colname which needs to plot on x-axis"
#data=df → The column Survived belongs to the DataFrame df
```

```python
df['Survived'].value_counts() #To get the freq of each category
```

Pie chart

```python
df['Survived'].value_counts().plot(kind ='pie')
```

PRACTICE ON SEX COLUMN

```python
df['Sex'].value_counts()
```

```python
df['Sex'].value_counts().plot(kind ='pie')
```

```python
sns.countplot(x='Sex', data=df)
```

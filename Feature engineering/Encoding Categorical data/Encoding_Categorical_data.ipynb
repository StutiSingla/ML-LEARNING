```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("customer.csv")
df.head()
```

1. Gender has (M,F) nomial categorical data - One Hot Encoder
2. review(G,P,Avg) ordinal categorical data - Ordinal Encoding
3. education (School,UG,PG) ordinal categorical data - Ordinal Encoding
4. purchased (Y/N) nomial categorical data - Label Encoding

As of now we will ignore age and Gender column

```python
df = df.iloc [:,2:]
df.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.iloc[:,0:-1], df.iloc[:,-1], test_size=0.2)
```

```python
from sklearn.preprocessing import OrdinalEncoder
oe = OrdinalEncoder(categories=[['Poor','Average','Good'],['School','UG','PG']])
oe.fit(X_train)
X_train = oe.transform(X_train)
X_test = oe.transform(X_test)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
le.fit(y_train)
y_train = le.transform(y_train)
y_test = le.transform(y_test)

y_train
```
